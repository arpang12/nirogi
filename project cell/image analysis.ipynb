```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

```python
r, g, b = cv2.split(nemo)
fig =plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
```

```python
pixel_colors = nemo.reshape((np.shape(nemo)[0]*np.shape(nemo)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
```

```python
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
##making color filter 
from matplotlib.colors import hsv_to_rgb
```

```python
light_green = (86, 245, 230)
dark_green = (111, 255.16, 230.00)
```

```python
>>> lo_square = np.full((10, 10, 3), light_green, dtype=np.uint8) / 255.0
>>> do_square = np.full((10, 10, 3), dark_green, dtype=np.uint8) / 255.0
```

```python
plt.subplot(1, 2, 1)
>>> plt.imshow(hsv_to_rgb(do_square))
>>> plt.subplot(1, 2, 2)
>>> plt.imshow(hsv_to_rgb(lo_square))
>>> plt.show()
```

```python
mask = cv2.inRange(dp, light_green, dark_green)
```

```python
result = cv2.bitwise_and(ds, ds, mask=mask)
```
